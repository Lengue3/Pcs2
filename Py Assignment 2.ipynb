```python
#Simple Array Sum

#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'simpleArraySum' function below.
#
# The function is expected to return an INTEGER.
# The function accepts INTEGER_ARRAY ar as parameter.
#

def simpleArraySum(ar):
    sm = 0
    for i in ar:
        sm = sm + int(i)
    return sm
    # Write your code here

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    ar_count = int(input().strip())

    ar = list(map(int, input().rstrip().split()))

    result = simpleArraySum(ar)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
#Compare the Triplets

#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'compareTriplets' function below.
#
# The function is expected to return an INTEGER_ARRAY.
# The function accepts following parameters:
#  1. INTEGER_ARRAY a
#  2. INTEGER_ARRAY b
#

def compareTriplets(a, b):
    alice = 0
    bob = 0
    for i in range(3):
        if a[i] > b[i]:
            alice = alice + 1
        elif a[i] < b[i]:
            bob = bob + 1
    return alice,bob
    # Write your code here

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    a = list(map(int, input().rstrip().split()))

    b = list(map(int, input().rstrip().split()))

    result = compareTriplets(a, b)

    fptr.write(' '.join(map(str, result)))
    fptr.write('\n')

    fptr.close()
```

```python
#Diagonal Difference

#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'diagonalDifference' function below.
#
# The function is expected to return an INTEGER.
# The function accepts 2D_INTEGER_ARRAY arr as parameter.
#

def diagonalDifference(arr,n):
    l_r = 0
    r_l=0
    for i in range(n):
        l_r = arr[i][i] + l_r
        r_l = arr[i][(n-1)-i] + r_l
    dd = abs(l_r-r_l)
    return dd
        
        
    # Write your code here

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    arr = []

    for _ in range(n):
        arr.append(list(map(int, input().rstrip().split())))

    result = diagonalDifference(arr,n)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
#Time Conversion

#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'timeConversion' function below.
#
# The function is expected to return a STRING.
# The function accepts STRING s as parameter.
#

def timeConversion(s):
    secs = []
    lst = s.split(":")
    h = lst[0]
    m = lst[1]
    s= lst[2]
    for i in s:
        secs.append(i)
    seconds = secs[0]+secs[1]
    am_pm = secs[2]+secs[3]
    if am_pm == "PM":
        hours = int(h)+12
        hours = str(hours)
        if hours == "24":
            hours = "12"
            
        result = f"{hours}:{m}:{seconds}"
    else:
        if h == "12":
            h = "00"
        result = f"{h}:{m}:{seconds}"
    return result
    # Write your code here

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    s = input()

    result = timeConversion(s)

    fptr.write(result + '\n')

    fptr.close()
```

```python
# Quicksort 1-Partition

def quickSort(arr):
    p = arr[0]
    left = []
    right=[]
    equal = []
    for i in arr:
        if i == p:
            equal.append(i)
        elif i > p:
            right.append(i)
        elif i< p:
            left.append(i)
    results = left+equal+right
    return results
```

```python
#Super Reducted String

def superReducedString(s):
    s = list(s)
    stack = []
    for i in s:
        if stack and stack[-1]==i:
            stack.pop()
        else:
            stack.append(i)
    if len(stack) == 0:
        return 'Empty String'
    else:
        return "".join(stack)
```

```python
#Caesar Cipher

def caesarCipher(s, k):
    og_ABC = list("ABCDEFGHIJKLMNOPQRSTUVWXYZ")
    og_abc = list("abcdefghijklmnopqrstuvwxyz")
    k_abc = list("abcdefghijklmnopqrstuvwxyz")
    k_ABC = list("ABCDEFGHIJKLMNOPQRSTUVWXYZ")
    cipher = []
    string = ""
    for i in range(k):
        old = k_abc[0]
        k_abc.remove(k_abc[0])
        k_abc.append(old)
    for _ in range(k):
        old = k_ABC[0]
        k_ABC.remove(k_ABC[0])
        k_ABC.append(old)
    for i in s:
        if i in og_ABC:
            pos = og_ABC.index(i)
            cipher.append(k_ABC[pos])
        elif i in og_abc:
            pos = og_abc.index(i)
            cipher.append(k_abc[pos])
        else:
            cipher.append(i)
    for l in cipher:
        string += l
    return string
```

```python
#Mars Exploration

def marsExploration(s):
    s = re.findall('...',s)
    counter = 0
    for i in s:
        if i != 'SOS':
            if i[0]!='S':
                counter += 1
            if i[1]!='O':
                counter += 1
            if i[2]!= 'S':
                counter += 1
    return counter
```
